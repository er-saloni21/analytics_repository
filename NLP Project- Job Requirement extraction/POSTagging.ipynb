```python
noun_phrase_array=[]
for values in df['job_description']:
  sen=sent_tokenize(values)
  for individual_sen in sen:
    chunk_line= nlp(individual_sen)
    for chunk in chunk_line.noun_chunks:
      noun_phrase_array.append(chunk)
```

```python
# initialise data of lists.
data = {'Noun_Phrases':noun_phrase_array}
 
# Create DataFrame
df_noun_phrase = pd.DataFrame(data)


print(df_noun_phrase)

```

```python
df_noun_phrase.to_csv('Noun_phrase_extraction.csv', index=False)
```

```python
len(df_noun_phrase)
```

the extracted noun phrases is not clean-- so we need to remove the unwanted characters, punctuation remove,
duplicate rows records remove

once that is done then we need to save it into csv

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english"))
clean_noun_phrase=[]
```

```python
for values in df_noun_phrase['Noun_Phrases']:
  val=values.text
  x= re.sub(r'[^A-Za-z ]+', '', val)
  x=x.lower()
  final=[]
  k=x.split(' ')
  text=''
  for word in k:
    if word not in stop_words:
      text = text + " " + word
      # final.append(word)

  
  clean_noun_phrase.append(text.strip())
  
```

```python
print(clean_noun_phrase)
```

```python
new_clean_noun_phrase = list(filter(None, clean_noun_phrase))
```
