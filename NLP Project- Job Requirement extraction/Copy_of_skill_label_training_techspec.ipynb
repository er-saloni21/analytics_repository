```python
import pandas as pd
import numpy as np
import io
```

```python
df= pd.read_csv(io.BytesIO(uploaded['training_noun_phrases.csv']), usecols=["training_noun_phrases"])
df.head()
```

```python
# Provide 'Address' as the column name
df['skill_set_bool'] = 0
```

```python
print(df)
```

```python
print(len(df))
```

```python
corpus = []
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
import re
```

```python
for i in range(0, len(df)):
    review = re.sub('[^a-zA-Z]', ' ', df['training_noun_phrases'][i])
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
print(corpus)
```

```python
df['training_noun_phrases']=corpus
```

```python
print(df)
```

```python
keyword_list = ['python','excel', 'database','queries','dashboards','plsql','modelling','tsql','integration','crm','salesforce','algorithm','impala','security','r','c','sas','bi','powerbi','power bi','powerpoint','pipeline','pandas','ssrs','ssis','alteryx','methodologies','visualization','powershell','sqldb','languages','qa','java','cloud','mssql','nosql','linux','dax','jira']

for index, row in df.iterrows():
    text_examine= row['training_noun_phrases'] 
    if set(keyword_list).intersection(text_examine.split()):
      df.loc[index, "skill_set_bool"]=1
```
