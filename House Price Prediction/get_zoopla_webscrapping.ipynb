```python
#from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
import requests

URL = https://www.zoopla.co.uk/for-sale/property/london/sw3/chelsea-brompton/?q=SW3&results_sort=newest_listings&search_source=home

res= requests.get(URL)

res.status_code

BeautifulSoup(res.text,'html.parser')

soups = BeautifulSoup(res.text,'html.parser')

last_item = int(soups.find_all("a", {"class":"css-1h66aar eaoxhri0"})[-2].text)
last_item

pages = []
zoopla_sale_properties=[]
for i in range(1, last_item):
    url = 'https://www.zoopla.co.uk/for-sale/property/london/sw3/chelsea-brompton/?page_size=25&q=SW7&radius=0&results_sort=newest_listings&pn=' + str(i)
    pages.append(url)
    
    
for item in pages:
    page = requests.get(item)
    soup = BeautifulSoup(page.text, 'html.parser')
    
    for page_item in soup.find_all(class_='earci3d1 css-fbt91f e2uk8e18'):
        stations=[]
        ports=[]
        price= page_item.find(class_='css-1smjogk eczcs4p0').text.replace('\n','').strip()
        house_type = page_item.find(class_='css-rqb6pt e2uk8e14').text.replace('\n','').strip()
    #bedrooms= item.find(class_='css-v7kkwq eczcs4p0').text.replace('\n','').strip()
    #bedrooms = item.find("span", {"data-testid": "bed"}).find_next('p').contents[0]
     
        if page_item.find("span", {"data-testid": "bed"}):
            bedrooms= page_item.find("span", {"data-testid": "bed"}).find_next('p').contents[0]
        else:
            bedrooms = 0
        
        if page_item.find("span", {"data-testid": "bath"}):
            bathrooms= page_item.find("span", {"data-testid": "bath"}).find_next('p').contents[0]
        else:
            bathrooms = 0
        
        if page_item.find("span", {"data-testid": "london_underground_station"}):
            for part in page_item.find_all("span", {"data-testid": "london_underground_station"}):
                data=part.find_next('p').contents[0]
                stations.append(data)
        #data= item.find("span", {"data-testid": "london_underground_station"}).find_next('p').contents[0]
        
        if page_item.find("span", {"data-testid": "uk_ferry_port"}):
            for part in page_item.find_all("span", {"data-testid": "uk_ferry_port"}):
                data=part.find_next('p').contents[0]
                ports.append(data)
    
    
        zoopla_sale_properties.append([price,house_type,bedrooms,bathrooms,stations,ports])
        print("Price:", price)
        print("house type:", house_type)
        print("bedrooms:",bedrooms)
        print("bathrooms", bathrooms)
        print("stations", stations)
        print("ports",ports)
        print("\n")



import pandas as pd
pd.DataFrame(zoopla_sale_properties, columns=['Price','House_Type','Bedrooms','Bathrooms','Stations','Ports']).to_csv("zoopla_sales.csv")
```
